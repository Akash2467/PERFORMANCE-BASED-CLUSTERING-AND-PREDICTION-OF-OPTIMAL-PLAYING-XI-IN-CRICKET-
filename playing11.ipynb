```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
t20_batting=pd.read_csv('D:\\Studies\\Bangalore\\Python\\datasets for project\\Cleaned dataset\\t20_batting_cleaned.csv')
t20_bowling=pd.read_csv('D:\\Studies\\Bangalore\\Python\\datasets for project\\Cleaned dataset\\t20_bowling_cleaned.csv')
t20_fielding=pd.read_csv('D:\\Studies\\Bangalore\\Python\\datasets for project\\Cleaned dataset\\t20_fieldling_cleaned.csv')
```

```python
#filter the dataset
t20battingf=t20_batting[t20_batting['Inns']>=10]
t20bowlingf=t20_bowling[t20_bowling['Inns']>=10]
t20fieldingf=t20_fielding[t20_fielding['Inns']>=10]
```

```python
#create new metrics for ranking
t20battingf['ConversionRate'] = (t20battingf['100'] + t20battingf['50']) / t20battingf['Inns']
t20battingf['DuckRate'] = (t20battingf['0'] / t20battingf['Inns']) * 100
```

```python
t20bowlingf['W/Inn']=(t20bowlingf['Wkts']/t20bowlingf['Inns'])
t20bowlingf['Impactspells']=((t20bowlingf['4'] + t20bowlingf['5'])/t20bowlingf['Inns'])
```

```python
t20bowlingf['bAve']=t20bowlingf['Ave']
t20bowlingf['bSR']=t20bowlingf['SR']
```

```python
t20bowlingf=t20bowlingf.drop(columns=['SR'])
t20bowlingf=t20bowlingf.drop(columns=['Ave'])
```

```python
t20bowlingf
```

```python
t20fieldingf['St/Inn']=(t20fieldingf['St']/t20fieldingf['Inns'])
t20fieldingf['Ct/Inn']=(t20fieldingf['Ct']/t20fieldingf['Inns'])
```

```python
#Merge the datas
batting_data=t20battingf
bowling_data=t20bowlingf
fielding_data=t20fieldingf
combined_data_t20 = pd.merge(batting_data, bowling_data, on='Name', how='outer', suffixes=('_batting', '_bowling'))
combined_data_t20 = pd.merge(combined_data_t20, fielding_data, on='Name', how='outer', suffixes=('', '_fielding'))
```

```python
combined_data_t20['Startyear'] = combined_data_t20[['Startyear_batting', 'Startyear_bowling', 'Startyear']].min(axis=1)
combined_data_t20['Endyear'] = combined_data_t20[['Endyear_batting', 'Endyear_bowling', 'Endyear']].max(axis=1)

# Drop redundant columns
columns_to_drop = ['Startyear_batting', 'Startyear_bowling', 'Endyear_batting', 'Endyear_bowling']
combined_data_t20.drop(columns=columns_to_drop, inplace=True)
```

```python
combined_data_t20=combined_data_t20.fillna(0)
```

```python
#define metrics used to calculate the score
metrics=['Ave','SR','ConversionRate','DuckRate','W/Inn','bAve','bSR','Impactspells','Econ','D/I','St/Inn','Ct/Inn']
```

```python
#Define the weights
weights = {
    'Inns':0.2
    'Ave': 0.3,
    'SR': 0.4,
    'ConversionRate': 0.1,
    'DuckRate': -0.1,  
    'W/Inn': 0.3,
    'Econ': -0.4, 
    'Impactspells': 0.3,
    'bAve':-0.4,
    'bSR':-0.4,
    'D/I': 0.5,
    'Ct/Inn':0.3,
    'St/Inn': 0.5
}
```

```python
#Calculate Batting score
combined_data_t20['BattingScore'] = (
    weights['Ave'] * combined_data_t20['Ave'] +
    weights['SR'] * combined_data_t20['SR'] +
    weights['ConversionRate'] * combined_data_t20['ConversionRate'] +
    weights['DuckRate'] * combined_data_t20['DuckRate']
)
```

```python
#Calculate bowling score
combined_data_t20['BowlingScore'] = (
    weights['W/Inn'] * combined_data_t20['W/Inn'] +
    weights['Econ'] * combined_data_t20['Econ'] +
    weights['Impactspells'] * combined_data_t20['Impactspells']+
    weights['bSR']*combined_data_t20['bSR']+
    weights['bAve']*combined_data_t20['bAve']
)
```

```python
combined_data_t20['FieldingScore'] = (
    weights['D/I'] * combined_data_t20['D/I'] +
    weights['Ct/Inn'] * combined_data_t20['Ct/Inn']+
    weights['St/Inn']*combined_data_t20['St/Inn']
)
```

```python
combined_data_t20.to_csv('D:\\Studies\\Bangalore\\Python\\datasets for project\\combined_t20_rank.csv')
```

```python
#combined_data_t20['BatRank']= combined_data_t20['BattingScore'].rank(ascending=False)
combined_data_t20['BatRank'] = combined_data_t20['BattingScore'].rank(ascending=False, method='dense').astype(int)
#combined_data_t20['BowlRank'] = combined_data_t20['BowlingScore'].rank(ascending=False)
combined_data_t20['BowlRank'] = combined_data_t20['BowlingScore'].rank(ascending=False, method='dense').astype(int)
#combined_data_t20['FRank'] = combined_data_t20['FieldingScore'].rank(ascending=False)
combined_data_t20['FRank'] = combined_data_t20['FieldingScore'].rank(ascending=False, method='dense').astype(int)
```
