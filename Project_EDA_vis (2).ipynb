```python
#What is the relationship between no. of innings and runs scored in each formats
plt.figure(figsize=(15, 5))
plt.subplots_adjust(wspace=0.3)  
plt.subplot(1, 3, 1)
sns.scatterplot(x='Inns',y='Runs',data=test_batting,color='red')
plt.xlabel("Innings played")
plt.ylabel('Runs scored')
plt.title("Test Format")
plt.subplot(1, 3, 2)
sns.scatterplot(x='Inns',y='Runs',data=odi_batting,color='pink')
plt.xlabel("Innings played")
plt.ylabel('Runs scored')
plt.title("ODI Format")
plt.subplot(1, 3, 3)
sns.scatterplot(x='Inns',y='Runs',data=t20_batting,color='purple')
plt.xlabel("Innings played")
plt.ylabel('Runs scored')
plt.title("T20 Format")
```

```python
#countries producing most players in each format
# Filter data for Test, ODI, and T20 batting based on a threshold of 30 players
test_batting_filtered = test_batting[test_batting['Country'].map(test_batting['Country'].value_counts()) >= 30]
odi_batting_filtered = odi_batting[odi_batting['Country'].map(odi_batting['Country'].value_counts()) >= 30]
t20_batting_filtered = t20_batting[t20_batting['Country'].map(t20_batting['Country'].value_counts()) >= 30]
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.countplot(data=test_batting_filtered, x='Country',palette='rocket', order=test_batting_filtered['Country'].value_counts().index)
plt.subplots_adjust(wspace=0.3, bottom=0.2)
plt.xticks(rotation=45, fontsize=9)
plt.xlabel('Country')
plt.ylabel('Counts')
plt.title("Test Format")
plt.subplot(1, 3, 2)
sns.countplot(data=odi_batting_filtered, x='Country',palette='viridis', order=odi_batting_filtered['Country'].value_counts().index)
plt.subplots_adjust(wspace=0.3, bottom=0.2)
plt.xticks(rotation=45, fontsize=9)
plt.xlabel("Country")
plt.ylabel('Counts')
plt.title("ODI Format")
plt.subplot(1, 3, 3)
sns.countplot(data=t20_batting_filtered, x='Country',palette='coolwarm', order=t20_batting_filtered['Country'].value_counts().index)
plt.subplots_adjust(wspace=0.3, bottom=0.2)
plt.xticks(rotation=45, fontsize=7)
plt.xlabel("Country")
plt.ylabel('Counts')
plt.title("T20 Format")

plt.tight_layout()
plt.show()

```
